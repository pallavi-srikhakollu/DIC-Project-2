```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
import pandas as pd
df = pd.read_csv("ny_swimming_14Apr1.csv")
url_list = df['url']
```

```python
all_text = ""
filename = "ny_swimming_"
file_counter = 0
for urli in url_list:
    
    if "baseball" not in urli and "mlb" not in urli:
        continue
    all_text = ""
    file_counter += 1
    response = requests.get(urli)
    doc = BeautifulSoup(response.text, 'html.parser')
    links = doc.find_all('p')
    
    for name in links:
            for x in name.find_all('a'):
                x.replaceWithChildren()
                
    for link in links:
        all_text+=link.text
    
    if all_text == "":
        continue
        
    curr_file_name = filename + str(file_counter) + ".txt"
    text_file = open(curr_file_name, "w")
    text_file.write( all_text)
    text_file.close()
    print("New file writing",file_counter)

#print(all_text)
```
