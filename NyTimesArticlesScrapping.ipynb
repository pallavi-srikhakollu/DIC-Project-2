```python
import pandas as pd
df = pd.read_csv("ny_nfl_3Apr6.csv")
url_list = df['url']
#url_lis = url_list[:10]
```

```python
all_text = ""
for urli in url_list:
    response = requests.get(urli)
    doc = BeautifulSoup(response.text, 'html.parser')
    links = doc.find_all('article')

    for link in links:
        all_text+=link.text
        #print(link.text)
print(all_text)
```

```python
text_file = open("Text3Apr6.txt", "w")
text_file.write( all_text)
text_file.close()
```
